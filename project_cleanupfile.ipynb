```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
file = 'Motor_Vehicle_Crashes_modified.csv'
```

```python
motor_vehicle_df = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
motor_vehicle_df.head()
#motorvehicle_df.head()
```

```python
del motor_vehicle_df['Unnamed: 12']
motor_vehicle_df.head()
```

```python
#motor_vehicle_df.count()
#motorvehicle_df.count()
motor_vehicle_df['Vehicle Make'].value_counts()
```

```python
#motor_vehicle_df['Vehicle Make'].replace('TYOYTA', 'TOYOTA', inplace = True)
motor_vehicle_df['Vehicle Make'] = motor_vehicle_df['Vehicle Make'].replace({'FRD':'FORD','NISSA': 'NISSAN','HYUND' :'HYUNDAI','SUBAR':'SUBARU'})

```

```python
motor_vehicle_df['Vehicle Make'].value_counts()
```

```python
motor_vehicle_df.count()
```

```python
motor_vehicle_df['Vehicle Make'].replace('', np.nan, inplace = True)
motor_vehicle_df['Vehicle Year'].replace('', np.nan, inplace = True)
motor_vehicle_df['Type / Axles of Truck or Bus'].replace('', np.nan, inplace = True)
motor_vehicle_df['State of Registration'].replace('', np.nan, inplace = True)
motor_vehicle_df['Contributing Factor 1'].replace('', np.nan, inplace = True)
motor_vehicle_df['Contributing Factor 1 Description'].replace('', np.nan, inplace = True)
motor_vehicle_df['Contributing Factor 2'].replace('', np.nan, inplace = True)
motor_vehicle_df['Contributing Factor 2 Description'].replace('', np.nan, inplace = True)
```
